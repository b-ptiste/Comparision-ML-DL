# usefull functions

This function allows you to save a JSON

```python
def save_dct_as_json(dct,name,path):
    os.chdir(path)
    with open(f'{name}.json', 'w') as fp:
        json.dump(dct, fp)
```

This function will be used for the display to make selections among the hyper-parameters of the gridsearch

```python
def word_in(x, pattern):
    if re.search(pattern, x, flags=0) is  None:
        return False
    else:
        return True
```

This function opens the Json of the SVM

```python
def open_json_svm():
    with open('svm.json') as f:
        map_score_svm_pca = json.load(f)
        
    svm_pca = list(map_score_svm_pca.keys())
    df_svm_pca = pd.DataFrame(index=svm_pca, data = map_score_svm_pca.values(), columns=["av. accuracy", "av. time"])
    
    return df_svm_pca.copy()
```

This function allows you to open the SVM Json with PCA

```python
def open_json_svm_pca(n):
    with open(f'svm_pca_{n}.json') as f:
        map_score_svm_pca = json.load(f)
        
    svm_pca = list(map_score_svm_pca.keys())
    df_svm_pca = pd.DataFrame(index=svm_pca, data = map_score_svm_pca.values(), columns=["av. accuracy", "av. time"])
    
    return df_svm_pca.copy()
```

This function opens the Json of the rf

```python
def open_json_rf():
    with open('rf.json') as f:
        map_score_svm_pca = json.load(f)
        
    svm_pca = list(map_score_svm_pca.keys())
    df_svm_pca = pd.DataFrame(index=svm_pca, data = map_score_svm_pca.values(), columns=["av. accuracy", "av. time"])
    
    return df_svm_pca.copy()
```

This is an interactive function that allows you to plot only the information you want. It is very useful for model selection and comparison.

```python
def plot_res(name_df, selections, compare):
    plt.figure(figsize=(30,10))
    for name in name_df:
        df = map_df[name]
        if selections is not None:
            for selection in selections:
                df = df[df.index.map(lambda x: word_in(x, selection))]
        n_row = df.shape[0]
        if "time" in compare:
            plt.plot(df.index, df["av. time"], label=f"{name} - av. time")
        if "accuracy" in compare:
            plt.plot(df.index, df["av. accuracy"], label=f"{name} - av. accuracy")
        
    plt.xticks(rotation=45)    
    plt.legend()
    plt.plot()
```

# SVM tunning

This function allows you to test several parameter combinations by doing a cross validation and it also returns the average execution time

```python
def svm_kernel_tunning(train_data, train_label, test_data, test_label, kernel, C, decision_function_shape, deg):
    cv = 5
    t1 = time.time()
    clf_svm = SVC(decision_function_shape=decision_function_shape, 
                  kernel=kernel, 
                  C=C,
                  degree = deg)
    scores = cross_val_score(clf_svm, train_data, train_label, cv = cv)
    return np.mean(scores), (time.time()-t1)/cv
```

Here is the list of parameters that we will study in exhaustive gridsearch

```python
list_kernel = ['linear', 'poly', 'rbf', 'sigmoid']
list_deg = [i for i in range(2,9,2)]
list_C = [10**i for i in range(-5,3)]
list_decision_function_shape = ["ovo", "ovr"]
```

This cell allows all combinations to be tested. Then the results are stored in a dictionary and saved in a JSON

```python
with_reload = False
if with_reload:
    map_score_svm = {}
    for kernel in tqdm(list_kernel):
        for C in list_C:
            for decision_function_shape in list_decision_function_shape:
                if kernel == "poly":
                    for deg in list_deg:
                        score, time_ex = svm_kernel_tunning(train_data, train_label, test_data, test_label, kernel, C, decision_function_shape, deg)
                        map_score_svm[f"{kernel} ({deg}) - {C} - {decision_function_shape}"] = score, time_ex

                else:
                    score, time_ex = svm_kernel_tunning(train_data, train_label, test_data, test_label, kernel, C, decision_function_shape, 0)
                    map_score_svm[f"{kernel} - {C} - {decision_function_shape}"] = score, time_ex

    save_dct_as_json(map_score_svm, "svm", os.getcwd())
```

This function allows you to do the PCA. Then calculate the percentage of variance and also have the cumulative percentage.

```python
pca = PCA()
pca.n_components = 784
pca.fit(train_data)
pca_train = pca.transform(train_data)
pca_test = pca.transform(test_data)
percentage_var_explained = pca.explained_variance_ / np.sum(pca.explained_variance_)
cum_var_explained = np.cumsum(percentage_var_explained)
```

Thus, we can approximate the cumulative variance

```python
plt.figure(1, figsize=(12, 8))
plt.clf()
plt.plot(cum_var_explained, linewidth=2)
plt.axis("tight")
plt.grid()
plt.xlabel("n_components")
plt.ylabel("Cumulative_explained_variance")
plt.show()
```

This function allows the gridsearch to be performed according to a certain amount of variance

```python
def grilled_search_pca(p, train_data, test_data):
    pca = PCA()
    pca.n_components = 784
    pca.fit(train_data)
    pca_train = pca.transform(train_data)
    pca_test = pca.transform(test_data)
    percentage_var_explained = pca.explained_variance_ / np.sum(pca.explained_variance_)
    cum_var_explained = np.cumsum(percentage_var_explained)

    n = np.argmin(cum_var_explained<p)
    pca_train_p = pca_train[:,:n]
    pca_test_p = pca_test[:,:n]

    map_score_svm_PCA = {}
    for kernel in tqdm(list_kernel):
        for C in list_C:
            for decision_function_shape in list_decision_function_shape:
                if kernel == "poly":
                    for deg in list_deg:
                        score, time_ex = svm_kernel_tunning(pca_train_p, train_label, pca_test_p, test_label, kernel, C, decision_function_shape, deg)
                        map_score_svm_PCA[f"{kernel} ({deg}) - {C} - {decision_function_shape}"] = score, time_ex

                else:
                    score, time_ex = svm_kernel_tunning(pca_train_p, train_label, pca_test_p, test_label, kernel, C, decision_function_shape, 0)
                    map_score_svm_PCA[f"{kernel} - {C} - {decision_function_shape}"] = score, time_ex
    
    save_dct_as_json(map_score_svm_PCA, f"svm_pca_{int(p*100)}", os.getcwd())
    return map_score_svm_PCA
```

we can test for several percentages of cumulative variance

This cell take about 2 hours

```python
with_reload = False
if with_reload:
    for p in [0.3, 0.6, 0.8, 0.9, 0.95, 0.99]:
        grilled_search_pca(p, train_data, test_data)
```

# Neural Network

This section focuses on a shallow CNN

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

labels = [i for i in range(10)]
num_train = 5000   # 60000 for full data set 
num_test  =  500   # 10000 for full data set

# Scale images to the [0, 1] range
x_train = x_train.astype("float32") / 255
x_test = x_test.astype("float32") / 255

num_classes = 10
input_shape = (28, 28, 1)

x_train = x_train[:num_train,:,:]
x_test = x_test[:num_test,:,:]

x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)

y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

defining the model with keras

First model with a single convolution layer

```python
model = keras.Sequential()

model.add(keras.Input(shape=input_shape))
model.add(Conv2D(32, kernel_size=(3, 3), activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation="softmax"))


model.summary()

# training 

batch_size = 128
epochs = 10

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

t1 = time.time()
hist = model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
t2 = time.time()-t1
print(t2)
```

```python
pd.DataFrame(hist.history)
pd.DataFrame(hist.history)[["loss","val_loss"]].plot()
pd.DataFrame(hist.history)[["accuracy","val_accuracy"]].plot()
```

Second model with two convolution layers

```python
model = keras.Sequential()

model.add(keras.Input(shape=input_shape))
model.add(Conv2D(32, kernel_size=(3, 3), activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, kernel_size=(3, 3), activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation="softmax"))


model.summary()

# training 

batch_size = 128
epochs = 10

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

t1 = time.time()
hist = model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
t2 = time.time()-t1
print(t2)
```

```python
pd.DataFrame(hist.history)
pd.DataFrame(hist.history)[["loss","val_loss"]].plot()
pd.DataFrame(hist.history)[["accuracy","val_accuracy"]].plot()
```
